```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.pairplot(iris)
```

```python
ir = sns.pairplot(iris)
ir.map_diag(sns.distplot) # graficos de distribucion
ir.map_upper(plt.scatter) # puntos
ir.map_lower(sns.kdeplot) # dist kernel
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
t = sns.FacetGrid(data=tips, col='time', row='smoker')
t.map(sns.distplot, 'total_bill')
```

```python
t = sns.FacetGrid(data=tips, col='time', row='smoker')
t.map(plt.scatter, 'total_bill', 'tip')
```
