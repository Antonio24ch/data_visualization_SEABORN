```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.lmplot(x= 'total_bill', y= 'tip', data=tips)
```

```python
sns.lmplot(x= 'total_bill', y= 'tip', data=tips, hue= 'smoker')
```

```python
sns.lmplot(x= 'total_bill', y= 'tip', data=tips, hue='smoker', markers= ['x','v'],
          scatter_kws={'s': 30})
```

```python
sns.lmplot(x= 'total_bill', y= 'tip', data=tips, hue='sex', markers= ['x','v'],
          scatter_kws={'s': 30})
```

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips, col='sex')
```

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips, col='smoker')
```

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips, col='sex', row= 'smoker')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect= 0.6, height= 5)
```
